```python
n = int(input("Введите количество учеников: "))
girl = int(input("Введите количество девочек: "))
boy = n-girl
boys_count, girls_count, boys_sum, girls_sum = 0, 0, 0, 0
for _ in range(girl):
    height = int(input("Введите рост ученика: "))
    if height >= 0:
        girls_count += 1
        girls_sum += height
for _ in range(boy):
    height = int(input("Введите рост ученика: "))
    if height >= 0:
        boys_count += 1
        boys_sum += height    
girls_avg_height = girls_sum / girls_count
boys_avg_height = boys_sum / boys_count
print("Средний рост девочек:", girls_avg_height)
print("Средний рост мальчиков:", boys_avg_height)
```
